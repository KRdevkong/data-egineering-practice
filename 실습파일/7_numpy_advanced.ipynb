```python
"""
    브로드캐스팅 규칙
규칙 1: 두 배열의 차원수가 다른 경우
더 작은 차원의 배열에 차원을 추가(형상 왼쪽에 1 추가)

규칙 2: 두 배열의 형상이 특정 차원에서 일치하지 않은 경우
해당 차원이 1인 배열을 다른 배열의 형상과 일치하도록 증가

규칙 3: 임의의 차원에서 크기가 일치하지않고 1도 아닌 경우
브로드 캐스팅 불가
"""
#1
a = np.ones((2,3))
b = np.arange(3)

a.shape ==(2,3)
b.shape ==(3, )
print(a + b)

#2
a = np.arange(3)
a = a.reshape((3,1))
b = np.arange(3)

a.shape ==(3, 1)
b.shape == (1, 3)

print(a + b)
# ex) 1D array + 2D array
```

```python
"""
    예제: 팬시 인덱싱을 이용한 임의의 점 집합 선택하기
"""

# 시각화 설정
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()  # set plot styles
%config InlineBackend.figure_format = 'svg'
```

```python
mean = [0, 0]     # 평균
cov = [[1, 2],    # 공분산
      [2, 5]]

# RandomState: 시드를 별도 객체에서 관리
rand = np.random.RandomState(1)

# multivariate_normal(): 정규분포 다변량 데이터 생성
x = rand.multivariate_normal(mean, cov, 100)

print(x[:10])
print(x.shape)
```

```python
# 산포도 출력: scatter(x, y)
plt.scatter(x[:, 0], x[:, 1]);
```

```python
# 20개의 점을 무작위 추출(random sampling) -> 인덱스 값을 리턴
ind = np.random.choice(x.shape[0],20,replace=False)
#x.shape = (100,2)
print(ind)
```

```python
# 선택된 점 집합 출력
selected = x[ind]

print(selected)
print(selected.shape)
```

```python
# 선택된 점 집합 시각화
plt.scatter(x[:, 0], x[:, 1], alpha=0.3)    # 원본 점 집합
plt.scatter(selected[:, 0], selected[:, 1], # 선택된 점 집합
            facecolor='red', s=50);
```
